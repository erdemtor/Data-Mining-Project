```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
data = pd.read_csv("/home/arikan/UT/Spring2017/DataMining/Data-Mining-Project/data.csv", encoding = "ISO-8859-1")
data["Lyrics"] = data.Lyrics.fillna("")
data["Artist"] = data.Artist.apply(lambda artist: re.sub(" and .*| featuring .*", '', artist))
artists_grouped = data.groupby('Artist').count()
popular_artists = data.Artist.value_counts()[:20]
```

```python
# number of songs per artist
fig = plt.figure(figsize=(10,8))
plt.hist(artists_grouped['Song'], bins=40)
plt.xlabel('Number of songs', fontsize=15)
plt.ylabel('Number of artists', fontsize=15)
fig.suptitle('Number of songs per artist', fontsize=20)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
popular_artists.plot(kind='bar') 
plt.xlabel('Artists', fontsize=15)
plt.ylabel('Number of songs', fontsize=15)
fig.suptitle('Number of songs, top 20 artists', fontsize=20)
plt.show()
```
