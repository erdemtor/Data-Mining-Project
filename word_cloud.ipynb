```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from wordcloud import WordCloud, STOPWORDS
```

```python
data = pd.read_csv("/home/arikan/UT/Spring2017/DataMining/Data-Mining-Project/data.csv", encoding = "ISO-8859-1")
data["decade"] = data.Year.apply(lambda x: (x - x % 10))
data["Lyrics"] = data.Lyrics.fillna("")
```

```python
def generate_wordcloud(decade, imagefile, outfile):
    text = ' '.join(data.Lyrics.tolist())
    alice_mask = np.array(Image.open(imagefile))
    stopwords = set(STOPWORDS)
    stopwords.add('im')
    stopwords.add('got')
    stopwords.add('oh oh')
    wc = WordCloud(background_color="white", max_words=300, mask=alice_mask, stopwords=stopwords)
    wc.generate(text)
    wc.to_file(outfile)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.figure()
    plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
generate_wordcloud(decade, 
                   '/home/arikan/UT/Spring2017/DataMining/Data-Mining-Project/xd.png', 
                   '/home/arikan/UT/Spring2017/DataMining/Data-Mining-Project/' + str('ALL') + '.png')
```
