```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("/home/arikan/UT/Spring2017/DataMining/Data-Mining-Project/data.csv", encoding = "ISO-8859-1")
data["Lyrics"] = data.Lyrics.fillna("")
data["CharCount"] = data.Lyrics.apply(lambda x: len(x))
data["WordCount"] = data.Lyrics.apply(lambda x: len(x.split(' ')))
data["UniqWordCount"] = data.Lyrics.apply(lambda x: len(set(x.split(' '))))
data = data[data['CharCount'] > 50]
```

```python
def plot(x, y, title, xlabel, ylabel):
    m, b = np.polyfit(x, y, 1)
    fig = plt.figure(figsize=(10,8))
    plt.plot(x, m*x + b, '-', c='r')
    plt.scatter(x, y, s=10, c='b', marker="D", alpha=0.5)
    fig.suptitle(title, fontsize=20)
    plt.xlabel(xlabel, fontsize=18)
    plt.ylabel(ylabel, fontsize=16)
    plt.show()
```

```python
plot(data.Year, data.CharCount, 'Number of characters per song', 'Years', 'Character count')
```

```python
plot(data.Year, data.WordCount, 'Number of words per song', 'Years', 'Word count')
```

```python
plot(data.Year, data.UniqWordCount, 'Number of unique words per song', 'Years', 'Unique word count')
```
